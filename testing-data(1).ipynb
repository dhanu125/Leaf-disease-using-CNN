```python
import PIL
from PIL import Image
import numpy as np
from keras import backend as K
import matplotlib.pyplot as plt
from keras.models import load_model
```

```python
model=load_model('../input/saved-model/Leaf_disease.h5')   
```

```python
imgpath='../input/testing/pepperbell1 (5).JPG'

img = Image.open(imgpath)
plt.imshow(img)
size = (64, 64)
img = img.resize(size, PIL.Image.ANTIALIAS)  
img_array = np.array(img)
if K.image_data_format() == "channels_last":
    img_array = np.reshape(img_array,[1,64,64,3])
else:
    img_array = np.reshape(img_array,[1,3,64,64])
img_array = img_array/255

prediction = model.predict_classes(img_array)
if prediction==0:
    print("Pepper__bell___Bacterial_spot\n")
elif prediction==1:
    print("Potato___healthy\n")
elif prediction==2:
    print("Tomato_Leaf_Mold\n")
elif prediction==3:
    print("Tomato__Tomato_YellowLeaf__Curl_Virus\n")
elif prediction==4:
    print("Tomato_Bacterial_spot\n")
elif prediction==5:
    print("Tomato_Septoria_leaf_spot\n")
elif prediction==6:
    print("Tomato_healthy\n")
elif prediction==7:
    print("Tomato_Spider_mites_Two_spotted_spider_mite\n")
elif prediction==8:
    print("Tomato_Early_blight\n")
elif prediction==9:
    print("Tomato__Target_Spot\n")
elif prediction==10:
    print("Pepper__bell___healthy\n")
elif prediction==11:
    print("Potato___Late_blight\n")
elif prediction==12:
    print("Tomato_Late_blight\n")
elif prediction==13:
    print("Potato___Early_blight\n")
else:
    print("Tomato__Tomato_mosaic_virus\n")
```
